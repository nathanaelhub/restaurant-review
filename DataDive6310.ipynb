```python

# Function to process text and extract trigrams
def extract_trigrams(texts):
    trigrams = []
    for text in texts:
        # Tokenize the text
        tokens = nltk.word_tokenize(text.lower())
        # Generate trigrams from the tokens
        trigrams.extend(list(ngrams(tokens, 3)))
    return trigrams

# Extract trigrams for chicken_1 and chicken_5
trigrams_1 = extract_trigrams(chicken_1['text'].dropna())
trigrams_5 = extract_trigrams(chicken_5_with_chicken['text'].dropna())

# Count frequencies of trigrams
counter_trigrams_1 = Counter(trigrams_1)
counter_trigrams_5 = Counter(trigrams_5)

# Create DataFrames from the trigram counters
df_trigram_1 = pd.DataFrame(counter_trigrams_1.most_common(), columns=['trigram_1', 'freq_1'])
df_trigram_5 = pd.DataFrame(counter_trigrams_5.most_common(), columns=['trigram_5', 'freq_5'])

# Concatenate the two DataFrames
df_trigrams = pd.concat([df_trigram_1, df_trigram_5], axis=1)

# Display the first few rows to inspect the combined DataFrame
df_trigrams.head()
```

```python
# Load stopwords and initialize lemmatizer
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

# Function to process text and generate n-grams
def process_text_and_ngrams(texts, n):
    tokens = [word for text in texts for word in word_tokenize(text.lower()) if word.isalpha() and word not in stop_words]
    lemmatized = [lemmatizer.lemmatize(token) for token in tokens]
    ngrams_generated = list(ngrams(lemmatized, n))
    return ngrams_generated

# Apply processing and calculate n-grams frequencies
def ngram_frequencies(reviews, n):
    ngrams_list = process_text_and_ngrams(reviews['text'].dropna(), n)
    ngram_counts = Counter(ngrams_list)
    return pd.DataFrame(ngram_counts.most_common(100), columns=['ngram', 'Frequency']).reset_index(drop=True)

# Data processing for chicken_1
bi_1 = ngram_frequencies(chicken_1, 2)
tri_1 = ngram_frequencies(chicken_1, 3)
quad_1 = ngram_frequencies(chicken_1, 4)

# Data processing for chicken_5
bi_5 = ngram_frequencies(chicken_5, 2)
tri_5 = ngram_frequencies(chicken_5, 3)
quad_5 = ngram_frequencies(chicken_5, 4)

# Rename columns for clarity

bi_1 = bi_1.rename(columns={'ngram': 'bigram', 'Frequency': 'bi_freq_1'})
tri_1 = tri_1.rename(columns={'ngram': 'trigram', 'Frequency': 'tri_freq_1'})
quad_1 = quad_1.rename(columns={'ngram': 'quadgram', 'Frequency': 'quad_freq_1'})

bi_5 = bi_5.rename(columns={'ngram': 'bigram', 'Frequency': 'bi_freq_5'})
tri_5 = tri_5.rename(columns={'ngram': 'trigram', 'Frequency': 'tri_freq_5'})
quad_5 = quad_5.rename(columns={'ngram': 'quadgram', 'Frequency': 'quad_freq_5'})

# Combine the tables for comparison
comparison_table = pd.concat([bi_1, tri_1, quad_1, bi_5, tri_5, quad_5], axis=1)

# Display the first 20 rows
comparison_table.head(20)
```

# Analysing results for both 1 and 5 star reviews

### Observations and Trends

 

#### There were many observations and trends that emerged from the N-Gram Data. First, with regard to recurring meals being ordered, common dishes such as "orange chicken," "chicken fried rice," and "sweet sour chicken" appear in both 1-star and 5-star reviews, suggesting these are frequently ordered when dinging and discussed in reviews. However, the frequency differences suggest that satisfaction with these dishes can vary significantly, based on the restaurant's execution of the recipe or other service factors. Second, with regard to dish-specific comments, N-grams like "ordered sweet sour chicken," "orange chicken fried rice," and "chicken fried rice chicken" suggest that specific preparation styles or combinations might be particularly memorable, either positively or negatively. The presence of specific dish mentions in 5-star reviews at higher frequencies (e.g., "orange chicken" appears more frequently in 5-star reviews) may suggest better satisfaction or notable quality when these dishes are part of a positive dining experience. Third, there are service-related mentions in the reviews. Phrases like "ordered orange chicken," "got sweet sour chicken," and similar phrases appearing in both 1-star and 5-star reviews hint at the importance of how the order fulfillment process impacts customer satisfaction. This can include speed, accuracy, and the condition of the dish upon serving – we lump these together under the umbrella term of customer service. 

 

### Indicators of 1-Star Reviews

 

#### If certain key dishes like "orange chicken" or "chicken fried rice" appear in 1-star reviews more than in other less negatively rated reviews, it indicates problems with these dishes, such as poor preparation or inconsistency. We saw this more frequently that customer service issues.

 

#### Customer service issues, however, were also mentioned in 1-star reviews. These included delays or mix-ups in service, as inferred from phrases like "ordered," might suggest service failures that contribute to lower ratings.

 

### Indicators of 5-Star Reviews

 

#### High frequencies of specific dishes in 5-star reviews suggest that these dishes are well-received when prepared to satisfaction, contributing significantly to positive experiences. Therefore, for Chinese and Mexican restaurants serving chicken-based meals, excellent food presentation/food quality is a strong indicator of a 5-star review. 

 

#### Positive customer service also lends itself to a 5-star review. Efficient and pleasing service, inferred from quick mentions of receiving orders like "got sweet sour chicken," could correlate with higher satisfaction and better ratings.

 

### General Thoughts and Future Recommendations

 

#### In our analysis, we narrowed our focus to Chinese and Mexican restaurants because the word “chicken” jumped out as a high frequency word in the 1-star reviews of all restaurants. When we dug deeper, we realized the top four categories of restaurants were Chinese-Restaurant, Restaurant-Chines, Mexican-Restaurant, and Restaurant-Chines. It is likely these two types of ethnic restaurants exist in higher quantities than other types of ethnic food categories. As such, they are also associated with common orders that include chicken. Comparing the trends we observed across different cuisines (Chinese vs. Mexican) that didn’t include chicken in the text review would be interesting as well as comparing all reviews that included the mention of chicken for both 1-star and 5-star.  Then, investigating whether certain themes are more prevalent in one cuisine over another could yield useful differentiation in customer expectations and satisfaction drivers.
